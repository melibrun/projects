3. проверяем на сайты рестрикции дальше
4. повторяем до тех пор пока сайтов рестрикции в последовательности не будет.

```python
import pandas as pd
```

```python
kodons = {"A": ["GCU","GCC","GCA","GCG"], 
"C":["UGU","UGC"],
"D":["GAU","GAC"],
"E":["GAA","GAG"],
"F":["UUU","UUC"],
"G":["GGU","GGC","GGA","GGG"],
"H":["CAU","CAC"],
"I":["AUU","AUC","AUA"],
"J":["AGA","AGG"],
"K":["AAA","AAG"],
"L":["CUU","CUC","CUA","CUG","UUA","UUG"],
"M":["AUG"],
"N":["AAU","AAC"],
"P":["CCU","CCC","CCA","CCG"],
"Q":["CAA","CAG"],
"R":["CGU","CGC","CGA","CGG","AGA","AGG"],
"S":["UCU","UCC","UCA","UCG","AGU","AGC"],
"T":["ACU","ACC","ACA","ACG"],
"V":["GUU","GUC","GUG","GUA"],
"W":["UGG"],
"Y":["UAU","UAC"]}
```

```python
magik = {}
for  value in kodons.values():
    for p in value:
        p = p.replace("U", "T")
        magik[p] = []
        for k in value:
            k = k.replace("U", "T")             
            magik[p].append(k)
magik
```

сайты рестрикции, которых не должно быть в последовательности:

```python
g =  ["TAGG", "GGATC", "GGTCTC", "CCAGAG", "GTTTAAAC", "CAAATTTG"]
```

```python
def chek(seq):
    for ch in g:
        if seq.find(ch) > -1:
            return 0
    return 1
        
def real_change(seq):
    t1 = ""
    vars = [""]
    for i in range(0,len(seq),3):
        kur = seq[i:i+3]
        v1 = []
        for elems in vars:
            for repl in magik[kur]:
                t = elems + repl
                if chek(t1 + t) == 1:
                    v1.append(t)
        v2 = []
        if len(v1[0]) > 6:
            t1 = t1 + v1[0][:3]
            for i in v1:
                if (i[3:] not in v2):
                    v2.append(i[3:])
        else:
            v2 = v1
                
        vars = v2
        
    if len(t1+vars[0]) != len(seq):
        return "error"
    else: 
        return t1 + vars[0]

                
```

```python
df = pd.read_csv(r"C:\Users\melib\Documents\ac_nt_db.csv")
df['seq_nuc'] = df['seq_nuc'].str.replace('U','T')
df["seq_nuc_optim"] = [real_change(x) if chek(x) == 0 else x for x in df["seq_nuc"]]
df
```
